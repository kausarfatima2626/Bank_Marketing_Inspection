```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types

df.info()
```

```python

# Summary statistics

df.describe()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
# Set the style of the plot
sns.set_theme(style="whitegrid")

# Create a count plot for the target variable 'y'
plt.figure(figsize=(6,4))
sns.countplot(x='y', data=df, palette='Set2')

# Add titles and labels
plt.title('Distribution of Target Variable (y)', fontsize=14)
plt.xlabel('Subscribed to Term Deposit? (yes/no)', fontsize=12)
plt.ylabel('Count of Customers', fontsize=12)

# Display the plot
plt.show()
```

```python

# Correlation matrix for numerical features
# Select only the numerical columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Set up the matplotlib figure size
plt.figure(figsize=(10, 8))

# Draw the heatmap using seaborn
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)

# Add title
plt.title('Correlation Matrix for Numerical Features', fontsize=16)

# Show the plot
plt.show()
```
